```python
def renombrar_columna(df: pd.DataFrame, viejo: str, nuevo: str) -> pd.DataFrame:
    """
    Renombra una columna en un DataFrame.

    Args:
        df (pd.DataFrame): El DataFrame donde renombrar.
        viejo (str): Nombre actual de la columna.
        nuevo (str): Nombre nuevo de la columna.

    Returns:
        pd.DataFrame: DataFrame con la columna renombrada.
    """
    return df.rename(columns={viejo: nuevo})
```

```python

def unir_partidos_y_jugadores(details: pd.DataFrame, games: pd.DataFrame) -> pd.DataFrame:
    """
    Une la tabla de detalles de jugadores con la tabla de partidos.
    
    Params:
        details: DataFrame con estadÃ­sticas de jugadores (tiene GAME_ID y TEAM_ID)
        games: DataFrame con estadÃ­sticas de partidos (tiene GAME_ID, HOME_TEAM_ID, AWAY_TEAM_ID)
    
    Returns:
        DataFrame combinado con info de partidos + jugadores,
        y una columna TEAM_TYPE que indica si es HOME o AWAY.
    """
    
    # Merge por GAME_ID
    merged = details.merge(
        games,
        on="GAME_ID",
        how="left"
    )
    
    # Identificar si el jugador es local o visitante
    merged["TEAM_TYPE"] = merged.apply(
        lambda row: "HOME" if row["TEAM_ID"] == row["HOME_TEAM_ID"] else "AWAY",
        axis=1
    )
    
    return merged
```

```python
def unir_con_info_equipos(df_final: pd.DataFrame, teams_info: pd.DataFrame) -> pd.DataFrame:
    """
    Une la info de equipos a cada jugador/partido, considerando el TEAM_ID.
    
    Args:
        df_final: DataFrame de jugadores + partidos
        teams_info: DataFrame de info de equipos (TEAM_ID, ABBREVIATION, CITY, ARENA, etc.)
    
    Returns:
        DataFrame combinado con info de equipos agregada.
    """
    
    # Merge principal: unir info de cada jugador segÃºn su TEAM_ID
    df = df_final.merge(
        teams_info,
        on='TEAM_ID',
        how='left',
        suffixes=('', '_team')
    )
    
    
    return df
```
